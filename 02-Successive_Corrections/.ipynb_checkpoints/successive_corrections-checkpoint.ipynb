```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
import math
```

```python
### Create function for Cressman Analysis ###
def cressman(R,d):
    if d<=R:
        w = (R**2 - d**2)/(R**2 + d**2)
    else:
        w = 0
    return w
```

```python
### Read in observations ###
initial = np.loadtxt("RAOBs_201903131200.txt", delimiter = ",", dtype  = str)
data = initial[:,1:].astype(float)
heights = data[:,2]
```

```python
### Set up analysis map with a 22x28 rectangular grid of points ###
x = np.arange(22)
y = np.arange(28)

delta = 1.27
x0 = 18.9
y0 = -6.3

x = x0+(x*delta)
y = y0 + (y*delta)

x_grid,y_grid = np.meshgrid(x,y)

# gird x y coords in m
x_grid = x_grid/100
y_grid = y_grid/100

m=1/15000000
tan = ((((x_grid/m)**2)+((y_grid/m)**2))**(1/2))/(6371000*(1+math.cos(math.radians(90-60))))
lat = np.degrees((np.pi/2)-(2*np.arctan(tan)))
lon= np.degrees(np.arctan(y_grid/x_grid)) - 115
```

```python
### convert obs lat/long to x,y ###
data_lat=data[:,0]
data_lon=data[:,1]

sigma = (1+np.sin(np.radians(60)))/(1+np.sin(np.radians(data_lat)))

data_x = sigma*6371000*np.cos(np.radians(data_lat))*np.cos(np.radians(data_lon+115)) / (15000000)
data_y = sigma*6371000*np.cos(np.radians(data_lat))*np.sin(np.radians(data_lon+115)) / (15000000)
```

```python
min_dist= []
for i in range(len(data_x)):

        distances = ( (data_x[i]-data_x)**2 + (data_y[i]-data_y)**2 )**0.5
        non_zero_dist = distances[distances != 0]
        min_dist.append(non_zero_dist.min())
```

```python
min_dist=np.array(min_dist)
d_min=min_dist.mean()
R1 = 4 * d_min
R2 = 2.5 * d_min
R3 = 1.5 * d_min
```

```python
dist2 = []
weights = []
for i in range(len(x_grid)):
    for j in range(len(x_grid[i])):
        distances = ( (x_grid[i,j]-data_x)**2 + (y_grid[i,j]-data_y)**2 )**0.5
        d = distances
        dist = {"i":i,"j":j, "d" : d}
        dist2.append(dist)

for k in range(len(dist2)):
    for l in range(len(d)):
        weights.append(cressman(R1,dist2[k]['d'][l]))
    
        
weights = np.array(weights).reshape(28,22,135)
```

```python
### Perform 500mb geopotential height analyses using a Cressman weighting Function###
#Use radii of influence 4, 2.5, 1.5 *dmin
analysis_f = []
for m in range(len(x_grid)):
    for n in range(len(x_grid[i])):
        num = sum(weights[m,n,:]*heights)
        den = sum(weights[m,n,:])
        analysis = num/den
        analysis_f.append(analysis)
analysis_f = np.array(analysis_f).reshape(28,22)
```

```python
### First analysis, no successive corrections ###
proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(4,4),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)

#plot analysis (MAY NEED TO CHANGE VARIABLE NAMES/INDICES)#
cs1 = ax1.contour(lon,lat,analysis_f,colors='maroon',levels=np.arange(0,8000,60),transform=ccrs.PlateCarree())
plt.clabel(cs1,levels=np.arange(0,8000,60))
plt.title("500mb analysis first pass")
#plt.savefig('first_pass')
plt.show()
```

```python
### Next Analyses ###
in_grid=[]
out_grid_idx=[]
in_grid_idx =[]
for i in range(len(data_x)):
    min_x = np.floor((data_x[i] - .189)/.0127)
    max_x = np.ceil((data_x[i] - .189)/.0127)
    min_y = np.floor((data_y[i] - (-.063))/.0127)
    max_y = np.ceil((data_y[i] - (-.063))/.0127)
    if (0 <= min_x <= 21) and  (0 <= max_x <= 21) and (0 <= min_y <= 27) and (0 <= max_y <= 27):
        min_x=min_x.astype(int)
        max_x=max_x.astype(int)
        min_y= min_y.astype(int)
        max_y= max_y.astype(int)
        mat = np.matrix([[1, x_grid[min_y,min_x], y_grid[min_y,min_x], x_grid[min_y,min_x] * y_grid[min_y,min_x]],
                         [1, x_grid[max_y,min_x], y_grid[max_y,min_x], x_grid[max_y,min_x] * y_grid[max_y,min_x]],
                          [1, x_grid[min_y,max_x], y_grid[min_y,max_x], x_grid[min_y,max_x] * y_grid[min_y,max_x]],
                           [1, x_grid[max_y,max_x], y_grid[max_y,max_x], x_grid[max_y,max_x] * y_grid[max_y,max_x]]])
        mat2 = np.matrix([[analysis_f[min_y,min_x]],
                 [analysis_f[max_y,min_x]],
                 [analysis_f[min_y,max_x]],
                 [analysis_f[max_y,max_x]]]) 

        mat_inv = np.linalg.inv(mat)
        result = np.dot(mat_inv, mat2)
        final = result[0] + result[1]*data_x[i] + result[2]*data_y[i] + result[3]*data_x[i]*data_y[i]
        in_grid.append(final)
        in_grid_idx.append(i)
    else:
        out_grid_idx.append(i)
    
out_grid_idx = np.array(out_grid_idx)
in_grid_idx  = np.array(in_grid_idx)
in_grid_val  = np.array(in_grid).reshape(len(in_grid))
```

```python
#estimating analaysis values at observation stations
d_x_out = data_x[out_grid_idx.astype(int)]
d_y_out = data_y[out_grid_idx.astype(int)]

dist3 = []
weights2 = []
for i in range(len(x_grid)):
    for j in range(len(x_grid[i])):
        distances = ( (x_grid[i,j]-d_x_out)**2 + (y_grid[i,j]-d_y_out)**2 )**0.5
        d = distances
        dist = {"i":i,"j":j, "d" : d}
        dist3.append(dist)

for k in range(len(dist3)):
    these_weights = np.zeros(len(out_grid_idx))
    for l in range(len(d)):
        these_weights[l] = cressman(R1,dist3[k]['d'][l])
        #weights2.append(cressman(R1,dist2[k]['d'][l]))
    weights2.append(these_weights)

weights2 = np.array(weights2)



analysis_f2 = []
for i in range(len(out_grid_idx)):
        num = sum(weights2[:,i]*analysis_f.reshape(616))
        den = sum(weights2[:,i])
        analysis = num/den
        analysis_f2.append(analysis)
out_grid_val = np.array(analysis_f2)



original_length = len(in_grid_idx) + len(out_grid_idx)
com_analysis_2 = np.empty(original_length)
com_analysis_2[in_grid_idx] = in_grid_val
com_analysis_2[out_grid_idx] = out_grid_val
```

```python
# analysis values at analysis points
dist2 = []
weights = []
for i in range(len(x_grid)):
    for j in range(len(x_grid[i])):
        distances = ( (x_grid[i,j]-data_x)**2 + (y_grid[i,j]-data_y)**2 )**0.5
        d = distances
        dist = {"i":i,"j":j, "d" : d}
        dist2.append(dist)

for k in range(len(dist2)):
    for l in range(len(d)):
        weights.append(cressman(R2,dist2[k]['d'][l]))
    
        
weights = np.array(weights).reshape(28,22,135)


analysis2_f = []
for m in range(len(x_grid)):
    for n in range(len(x_grid[i])):
        num = np.nansum(weights[m,n,:]*((heights-com_analysis_2)))
        den = np.sum(weights[m,n,:])
        comb = num/den
        analysis2 = analysis_f[m,n] + comb
        analysis2_f.append(analysis2)
analysis2_f = np.array(analysis2_f).reshape(28,22)

```

```python
### Second analysis ###
proj = ccrs.Stereographic(central_longitude=-115,central_latitude=90,true_scale_latitude=60)
fig = plt.figure(figsize=(4,4),dpi=200)
ax1 = fig.add_subplot(111,projection=proj)
ax1.add_feature(cfeature.STATES)
ax1.add_feature(cfeature.COASTLINE)

#plot analysis (MAY NEED TO CHANGE VARIABLE NAMES/INDICES)#
cs1 = ax1.contour(lon,lat,analysis2_f,colors='maroon',levels=np.arange(0,8000,60),transform=ccrs.PlateCarree())
plt.clabel(cs1,levels=np.arange(0,8000,60))
plt.title("500mb analysis second pass")
#plt.savefig('second_pass')
plt.show()
```
