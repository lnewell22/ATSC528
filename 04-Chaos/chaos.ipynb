```python
### Import Required Modules (shouldn't need to change) ###
import numpy as np                 #numpy for math
import matplotlib.pyplot as plt    #matplotlib for plotting
import cartopy.crs as ccrs         #cartopy for plotting on map
import cartopy.feature as cfeature #cartopy basic shapefiles
```

```python
a = 0.25
b = 4.0
F = 8.0
G = 1.0

(x0,y0,z0) = (2.5, 1.0, 0.0)
(x01,y01,z01) = (2.49, 1.0, 0.0)
```

```python
time_step = 4/(24*5)
time_units = 120/5 #normalized time
times = np.arange(0,24.01,time_step)
```

```python
def n_t(x, y, z, time_diff):
    x_n = x + time_diff*(-(y**2) - (z**2) - a*x + a*F)
    y_n = y + time_diff*(x*y - b*x*z - y + G)
    z_n = z + time_diff*(b*x*y + x*z - z)
    return (x_n, y_n, z_n)
```

```python
(x,y,z) = (x0,y0,z0)
sim0 = np.zeros((len(times),3))
for idx in range (1,len(times)):
    (x,y,z) = n_t(x,y,z,time_step)
    sim0[idx] = (x,y,z)


(x1,y1,z1) = (x01,y01,z01)
sim1 = np.zeros((len(times),3))
for idx in range(1,len(times)):
    (x1,y1,z1) = n_t(x1,y1,z1,time_step)
    sim1[idx] = (x1,y1,z1)
```

```python
times2 = times*5
```

```python
ig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 10))
ax1.plot(times2,sim0[:,0],label='Sim 1')
ax1.plot(times2,sim1[:,0], label = 'Sim 2')
ax1.set_ylabel('X Values')
minor_ticks = np.arange(-3, 3, 0.5)
ax1.set_yticks(minor_ticks)
ax1.grid(which='both', alpha=0.3)
ax1.legend()


ax2.plot(times2,sim0[:,1],label='Sim 1')
ax2.plot(times2,sim1[:,1],label = 'Sim 2')
ax2.set_ylabel('Y Values')
minor_ticks = np.arange(-3, 3, 0.5)
ax2.set_yticks(minor_ticks)
ax2.grid(which='both', alpha=0.3)
ax2.legend()


ax3.plot(times2,sim0[:,2],label='Sim 1')
ax3.plot(times2,sim1[:,2],label = 'Sim 2')
ax3.set_ylabel('Z Values')
ax3.set_xlabel('Time (Days)')
minor_ticks = np.arange(-3, 3, 0.5)
ax3.set_yticks(minor_ticks)
ax3.grid(which='both', alpha=0.3)
ax3.legend()
```
